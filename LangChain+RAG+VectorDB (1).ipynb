```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings

from langchain.vectorstores import FAISS
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI

import os
```

```python
loader=TextLoader("/content/bigdata.txt")
documents=loader.load()
```

```python
splitter=CharacterTextSplitter(chunk_size=500,chunk_overlap=50)#chunk_overlap is used to not miss the context of the sentence
docs=splitter.split_documents(documents)
```

```python
embeddings=HuggingFaceEmbeddings()
vectorstore=FAISS.from_documents(docs,embeddings)
```

```python
os.environ["GOOGLE_API_KEY"]="AIzaSyBMHE6IArarAw4jNFe_ii1mR-qlEpnZWzo"
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-1.5-flash",temperature=0)# temperature-llm creativity level
```

```python
#langchain
qa_chain=RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever()#we can use it anywhere like a model file(pickle,hd5)
)
```

```python
question="What is BigData"
result=qa_chain.run(question)
print("Answer:",result)
```
